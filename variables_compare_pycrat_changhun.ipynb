```python
import pandas as pd
from itertools import *
from pycaret.regression import *
```

```python
dataset2 = pd.read_csv('./hospital_male_0802_test.csv', encoding='cp949')
dataset2.mm = dataset2.mm.astype('int64')
dataset2 = dataset2[~dataset2['tma'].astype(str).str.contains('2011-|2016-')].reset_index(drop=True)
# tmp_df = dataset2.shift(periods=527)
# tmp_df
```

```python

fixed_df = dataset2[['area', 'mm', 'weekday', 'frequency','heat_wave', 'cold_wave','total_pop', 'elderly_ratio','pm10','diff_hum','diff_temp','avg_ps', 'so2']]

sub_df = [ dataset2[['o3']], dataset2[['avg_age']], dataset2[['avg_temp']], dataset2[['min_max_ps']], dataset2[['sum_rn']], dataset2[['avg_hum']] ]
sub_list = list(combinations(sub_df, 2))
sub_list[2][0]
number = 1

for i in range(len(sub_df)):
    sub_list = list(combinations(sub_df, i+1))
    for j in range(len(sub_list)):
        print('number: ', number)
        dataset_df = fixed_df.copy()
        for k in sub_list[j]:
            dataset_df = pd.concat([dataset_df,k], axis=1)
        
        print(dataset_df.columns)
        data = dataset_df.sample(frac=0.9, random_state=786) 
        data_unseen = dataset_df.drop(data.index) 

        data.reset_index(drop=True, inplace=True) 
        data_unseen.reset_index(drop=True, inplace=True) 

        print('Data for Modeling: ' + str(data.shape)) 
        print('Unseen Data For Predictions: ' + str(data_unseen.shape)) 
        
        s = setup(data, target = 'frequency', session_id=123)
        
        best_model_5 = compare_models(sort='RMSE', n_select = 5)
        
        final_model = finalize_model(best_model_5[0])
        prediction = predict_model(final_model, data = data_unseen)
        print(prediction['Label'])
        print('='*100)
        number += 1
            
```
