## 기압

```python
# 지역, 시간별로 평균
ap2 = ap.drop(columns=['stn_id', 'yyyy','mm','dd']).groupby(['area', 'tma'], as_index = False).mean()
ap2
```

```python
ap2.isnull().sum()
```

#### min만 결측치일 때

```python
# min만 결측치일 때
t = ap2.loc[ap2['min_ps'].isnull() & ~ap2['max_ps'].isnull() & ~ap2['avg_ps'].isnull()]
x = t.index
```

```python
ap2.loc[ap2['min_ps'].isnull() & ~ap2['max_ps'].isnull() & ~ap2['avg_ps'].isnull()]
```

```python
for i in range(len(x)):
    ap2.loc[[x[i]], 'min_ps'] = ap2.loc[[x[i]], 'avg_ps'].to_list()[0] - ( ap2.loc[[x[i]], 'max_ps'].to_list()[0] - ap2.loc[[x[i]], 'avg_ps'].to_list()[0] )
```

```python
ap2.isnull().sum()
```

#### max만 결측치일 때

```python
# avg만 결측치일 때
t = ap2.loc[~ap2['min_ps'].isnull() & ~ap2['max_ps'].isnull() & ap2['avg_ps'].isnull()]
x = t.index
```

```python
t
```

```python
for i in range(len(x)):
    ap2.loc[[x[i]], 'avg_ps'] = ( ap2.loc[[x[i]], 'max_ps'].to_list()[0] + ap2.loc[[x[i]], 'min_ps'].to_list()[0] ) / 2
    
```

```python
ap2.isnull().sum()
```

#### 전부 결측치일 때

```python
# 전부 결측치일 때
t = ap2.loc[ap2['min_ps'].isnull() & ap2['max_ps'].isnull() & ap2['avg_ps'].isnull()]
x = t['tma'].to_list()
```

```python
t
```

```python
t.area.unique()
```

```python
for i in range(len(x)):
    ap2.loc[(ap2['tma']==x[i]) & (ap2['area']=='세종'), 'min_ps'] = ap2.loc[(ap2['tma']==x[i]) & (ap2['area']=='대전'), 'min_ps'].to_list()[0]
    ap2.loc[(ap2['tma']==x[i]) & (ap2['area']=='세종'), 'max_ps'] = ap2.loc[(ap2['tma']==x[i]) & (ap2['area']=='대전'), 'max_ps'].to_list()[0]
    ap2.loc[(ap2['tma']==x[i]) & (ap2['area']=='세종'), 'avg_ps'] = ap2.loc[(ap2['tma']==x[i]) & (ap2['area']=='대전'), 'avg_ps'].to_list()[0]
        
```

```python
ap2.isnull().sum()
```

```python
ap2.to_csv('./airPressure_edited_0714.csv', encoding='cp949')
```
