```python
import pandas as pd
import glob
```

### edited_data dir 에 있는 파일 작업 후 동일 dir, 동일 filename 으로 그대로 저장
- tma 삭제X > 추후 확인하고 삭제해도 될 듯해서 일단 삭제하지 않고 그대로 열 추가 합니다(yyyy, mm, dd)
- 지역명(area) 컬럼 맨 앞으로 이동

edited_data file list

```python
files = glob.glob('Z:/공모전/날씨빅데이터콘테스트(공유)/data/edited_data/*.csv')
files
```

```python
for file in files :
    data = pd.read_csv(file, encoding='cp949', index_col = 0)
    
    # tma 컬럼 data type 변경
    print('before data type - :', data.tma.dtype)
    data.tma = pd.to_datetime(data.tma)
    print('after data type  - :', data.tma.dtype)
    
    # yyyy, mm, dd : tma 뒤로 파생변수 생성(tmp drop X)
    data.insert(1, 'yyyy', data.tma.dt.year)
    data.insert(2, 'mm',   data.tma.dt.month)
    data.insert(3, 'dd',   data.tma.dt.day)
    
    # area col 위치 맨 앞으로 변경
    tmp = data.area
    data.drop(columns = 'area', axis=1, inplace=True)
    data.insert(0,'area', tmp)
    
    # 그대로 다시 저장
    display(data.head(3))
    data.to_csv(file, encoding='cp949', index=False)
```
