```python
# hum_df = pd.DataFrame(columns=['day','hour','forecast','습도','yyyymmdd','yyyy','mm','area'])
# temp_df = pd.DataFrame(columns=['day','hour','forecast','기온','yyyymmdd','yyyy','mm','area'])
# tot_df = pd.DataFrame(columns=['day','hour','forecast','value','yyyymmdd','yyyy','mm','area'])
hum_df = pd.DataFrame(columns=['day','hour','forecast','습도','yyyymmdd','yyyy','mm','area'])

# tot_df.rename(columns = {'value' : file_list_py[0].split('/')[-1].split('_')[-3]}, inplace=True)

display(hum_df)
first_path = './raw_data/2-2bigdata/*'
years = glob.glob(first_path)

for year in years:
    second_path = year + '/*'
    areas = glob.glob(second_path)
    for area in areas:
        path = area + '/*.csv'
        file_list = glob.glob(path)
        file_list2 = []
        for i in file_list:
            i = unicodedata.normalize('NFC', i)
            file_list2.append(i)
        file_list_py = [file for file in file_list2 if "습도" in file]

        print ("file_list_py: {}".format(file_list_py))

        for path in file_list_py:
            print(path)
            df = pd.read_csv(path, header=None)
            df.columns = ['day','hour','forecast','value']
            df.loc[0, 'yyyymmdd'] = df.loc[0,'value'].strip()[-8:]
            df.loc[df['day'].str.strip().str.startswith('Start'), 'yyyymmdd'] = df.loc[df['day'].str.strip().str.startswith('Start'), 'day'].str.strip().str[-8:]
            df.yyyymmdd = df.yyyymmdd.fillna(method='ffill')
            df.yyyymmdd = pd.to_datetime(df.yyyymmdd)
            
            df['area'] = path.split('/')[-2]
            df.rename(columns = {'value' : path.split('/')[-1].split('_')[-3]}, inplace=True)

            drop_idx = df[(df['hour'].isna()) & (df['forecast'].isna()) ].index
            df = df.drop(drop_idx).drop(0)
            
            try:
                df = df.iloc[df.index < df.index[df['day'].str.contains('format')][0]]
            except:
                None
                
            test = pd.DataFrame(( (df['day'].astype(int) * 24) + df['hour'].astype(int) / 100 ) + 9 + df['forecast'].astype(int))
            df.reset_index(drop=True, inplace=True)
            test.reset_index(drop=True, inplace=True)
            for i in range(0, len(df)):
                df.loc[i, 'yyyymmdd2'] = df.loc[i, 'yyyymmdd'] + datetime.timedelta(hours=test.loc[i][0]) - datetime.timedelta(days=1)
            
            df['yyyy']     = df.yyyymmdd2.dt.year
            df['mm']       = df.yyyymmdd2.dt.month
            df['day2']     = df.yyyymmdd2.dt.day
            display(df)
            hum_df = pd.concat([hum_df, df])
```
