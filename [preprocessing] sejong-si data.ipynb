### 합칠 세종 데이터 만들기
1. 2016년도 세종시 데이터 가져오기

```python
sj_2016 = glob.glob('/content/drive/MyDrive/공모전/날씨 빅데이터 콘테스트/data/airpollutants/2016/*.csv')
sj_2016
```

```python
data_2016 = pd.DataFrame()

for path in sj_2016 :
    frm = pd.read_csv(path, encoding='cp949')
    tmp = frm[frm['지역'].str.contains('세종')]
    data_2016 = data_2016.append(tmp, ignore_index = True)
print('complete -')
```

```python
# 쓸 컬럼만 가져오기
data2016 = data_2016[['지역', '측정일시','SO2', 'O3','PM10']]
data2016.columns = ['region','yyyymmdd','so2','o3','pm10']
data2016.shape
```

```python
# data2016.region.unique()  : '세종'만 있음
data2016.yyyymmdd = data2016.yyyymmdd.astype(str).str[:-2]

# groupby
data2016_ = data2016.groupby(['region','yyyymmdd'], as_index=False).mean()

data2016_.insert(2,'yyyy', pd.to_datetime(data2016_.yyyymmdd).dt.year)
data2016_.insert(3,'mm', pd.to_datetime(data2016_.yyyymmdd).dt.month)
data2016_.insert(4,'dd', pd.to_datetime(data2016_.yyyymmdd).dt.day)
data2016_
```

2. 2011.12 - 2015.12 세종시 데이터 만들기
- 대전 유성구

2011.12 데이터의 경우 한달만 필요하기 때문에 따로 처리함

```python
df = pd.read_excel('/content/drive/MyDrive/공모전/날씨 빅데이터 콘테스트/data/airpollutants/2011/2011년04분기.xlsx')
df_2011_4 = df.copy()
```

```python
tmp_df = df_2011_4[(df_2011_4['지역'] == '대전 유성구') & (df_2011_4['측정일시'].astype(str).str[4:6] == '12')]
tmp_df.drop(columns = ['측정소명','측정소코드','CO','NO2','주소'], inplace=True)
tmp_df.loc[:,'지역'] = '세종'
tmp_df.loc[:,'측정일시'] = tmp_df['측정일시'].astype(str).str[:-2]
tmp_df.rename(columns = {'지역':'region', '측정일시':'yyyymmdd','SO2':'so2','O3':'o3','PM10':'pm10'}, inplace=True)
tmp_df = tmp_df.groupby(['region','yyyymmdd'], as_index=False).mean()
tmp_df.insert(2,'yyyy', pd.to_datetime(tmp_df.yyyymmdd).dt.year)
tmp_df.insert(3,'mm', pd.to_datetime(tmp_df.yyyymmdd).dt.month)
tmp_df.insert(4,'dd', pd.to_datetime(tmp_df.yyyymmdd).dt.day)
print(tmp_df.shape)
tmp_df.head()
```

```python
sejong_air = tmp_df.append(data2016_)
sejong_air
```

현재 df
- data2016_ : 2016년도 세종시 데이터
- tmp_df : 2011년 12월 대전 유성구 데이터
- sejong_air == data2016_ + tmp_df
    - sejong_air 에 append

확장자 다르므로 다르게 진행

```python
path_lst = glob.glob('/content/drive/MyDrive/공모전/날씨 빅데이터 콘테스트/data/airpollutants/*')
path_lst = path_lst[:-2]
path_lst
```

```python
num = 1
for file in path_lst :
    # 확장자마다 다르게 데이터 로드
    try :
        if file.endswith('.xlsx') :
            df_ = pd.read_excel(file)
            data = df_.copy()

        elif file.endswith('.csv') :
            df_ = pd.read_csv(file, encoding='cp949') # error 생겨서 utf-8 로 except 문 추가...
            data = df_.copy()
    
    except :
        if file.endswith('.xlsx') :
            df_ = pd.read_excel(file)
            data = df_.copy()

        elif file.endswith('.csv') :
            df_ = pd.read_csv(file, encoding='utf-8')
            data = df_.copy()

    # 이후 과정은 동일
    data_ = data[data['지역'] == '대전 유성구']
    data_.drop(columns = ['측정소명','측정소코드','CO','NO2', '주소'], inplace=True)
    if 'PM25' in data_.columns.tolist():
        data_.drop(columns = 'PM25', inplace=True)
        
    data_.loc[:,'지역'] = '세종'
    data_.loc[:,'측정일시'] = data_['측정일시'].astype(str).str[:-2]
    data_.rename(columns = {'지역':'region', '측정일시':'yyyymmdd','SO2':'so2','O3':'o3','PM10':'pm10'}, inplace=True)
    data_ = data_.groupby(['region','yyyymmdd'], as_index=False).mean()
    data_.insert(2,'yyyy', pd.to_datetime(data_.yyyymmdd).dt.year)
    data_.insert(3,'mm', pd.to_datetime(data_.yyyymmdd).dt.month)
    data_.insert(4,'dd', pd.to_datetime(data_.yyyymmdd).dt.day)

    sejong_air = sejong_air.append(data_)
    print('dataframe size : ', sejong_air.shape)
    print(num, 'set complete!' )
    print('*' * 50)
    num += 1
```

```python
print('>>>> 세종시 데이터 - ')
display(sejong_air)
sejong_air.to_csv('/content/drive/MyDrive/공모전/날씨 빅데이터 콘테스트/data/airpollutants/세종시_airpollutants.csv', encoding='cp949', index=False)

df3_ = df3.append(sejong_air)
df3_ = df3_.sort_values(['region','yyyymmdd'])

print('\n>>>> 최종 데이터셋 확인 - ')
display(df3_)

print('\n>>>> 세종시 데이터 추가 확인 - ')
display(df3_[df3_.region == '세종']) # 세종시 데이터 추가 확인
df3_.to_csv('/content/drive/MyDrive/공모전/날씨 빅데이터 콘테스트/data/airpollutants/201112_201612_airpollutants_daily_modified.csv', encoding='cp949', index=False)
```
