1. 평균연령 데이터 처리

2. 노령인구 데이터 처리
- 2011.12 ~ 2012.06 까지 데이터 대전 데이터로 가져오기
- row 아예 존재하지 않으므로 대전 데이터 복사 후 그대로 이름만 변경

```python
elderly = pd.read_csv('Z:/공모전/날씨빅데이터콘테스트(공유)/data/201112_201612_성별_고령인구.csv', encoding='cp949')
```

```python
elderly.loc[(elderly.region == '세종'), :] # 2012-07-01 부터 데이터 존재
```

```python
tmp_sejong = elderly[elderly.region == '대전'].iloc[:14,:].copy()
tmp_sejong.region = '세종'
elderly = elderly.append(tmp_sejong)
elderly.info()
```

```python
elderly.sort_values(by=['region','yyyy','mm'], inplace=True)
```

```python
elderly.loc[(elderly.region == '세종'), :]
```

```python
elderly.to_csv('Z:/공모전/날씨빅데이터콘테스트(공유)/data/201112_201612_성별_고령인구.csv', encoding='cp949', index=False)
```
