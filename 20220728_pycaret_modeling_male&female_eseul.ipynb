```python
male_df = pd.read_csv('Z:/공모전/날씨빅데이터콘테스트(공유)/data/hospital_male_0725.csv', encoding='cp949')
male_df.mm = male_df.mm.astype('int64')
male_df.head()
```

```python
female_df = pd.read_csv('Z:/공모전/날씨빅데이터콘테스트(공유)/data/hospital_female_0725.csv', encoding='cp949')
female_df.mm = female_df.mm.astype('int64')
female_df.head()
```

```python
cols_1 = ['area','mm','weekday','total_pop','elderly_ratio',
          'so2','pm10','rhm_min_avg_dif','min_max_Ta','avg_ps',
          'o3','avg_ta','sum_rn', 'frequency']

cols_2 = ['area','mm','weekday','total_pop','elderly_ratio',
          'so2','pm10','rhm_min_avg_dif','min_max_Ta','avg_ps',
          'avg_age','sum_gsr','sum_rn','avg_rhm', 'frequency']

cols_3 = ['area', 'mm', 'weekday', 'total_pop', 'elderly_ratio',
        'so2', 'pm10', 'rhm_min_avg_dif', 'min_max_Ta', 'avg_ps',
        'avg_age', 'avg_ta', 'sum_gsr', 'sum_rn', 'avg_rhm', 'frequency']

female_dataset = female_df[cols_3]
female_dataset.head(3)
```

```python
male_data = male_dataset.sample(frac=0.9, random_state=786) 
male_data_unseen = male_dataset.drop(male_data.index) 

male_data.reset_index(drop=True, inplace=True) 
male_data_unseen.reset_index(drop=True, inplace=True) 

print('Data for Modeling: ' + str(male_data.shape)) 
print('Unseen Data For Predictions: ' + str(male_data_unseen.shape)) 
```

```python
female_data = female_dataset.sample(frac=0.9, random_state=786) 
female_data_unseen = female_dataset.drop(female_data.index) 

female_data.reset_index(drop=True, inplace=True) 
female_data_unseen.reset_index(drop=True, inplace=True) 

print('Female Data for Modeling: ' + str(female_data.shape)) 
print('Unseen Data For Predictions: ' + str(female_data_unseen.shape)) 
```

```python
# s = setup(male_data, target = 'frequency', session_id=123)
s = setup(female_data, target = 'frequency', session_id=123)
```

```python
best_model_5 = compare_models(sort='RMSE', n_select = 5)

final_model = finalize_model(best_model_5[0])
prediction = predict_model(final_model, data=female_data_unseen)
prediction.head()
```

```python
tuned_top5 = [tune_model(i, optimize = 'RMSE', n_iter=10) for i in best_model_5]

final_model = finalize_model(tuned_top5[0])
prediction = predict_model(final_model, data=female_data_unseen)
prediction.head()
```
