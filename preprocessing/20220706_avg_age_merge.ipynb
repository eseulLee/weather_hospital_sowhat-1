```python
import pandas as pd
import os
import glob
```

```python
path = './raw_data/201112_201612_avgAge.csv'
path
```

```python
avg_age_frm = pd.DataFrame(columns = ['region','yyyy','mm','sex','avg_age'])
```

1. 전국 삭제
2. column 명이 다 'yyyy월mm월남자(여자)평균연령' 으로 일치 하니까 전체 평균연령인 '평균연령' 으로 들어가는 컬럼 삭제
3. 위에 컬럼을 unstack? 해서 한 열로 싹 세우고 split 해서 yyyy, mm, sex(1,2) 열 만들기
4. 남/여 > 1/2 로 replace

```python
test = pd.read_csv(path, encoding='cp949', index_col=0)[1:]
```

```python
# 전체 평균연령 삭제
cols = [c for c in test.columns if c.find('월_평균연령') != -1]
```

```python
test = test.drop(columns=cols)
```

```python
# 각 컬럼을 돌면서
for cols in test.columns:
    #각 컬럼의 행을 돌면서 데이터 추가
    for rows in test[[cols]].index:
        yyyy = cols.split('_')[0][:4]
        mm   = cols.split('_')[0][5:7]
        sex  = cols.split('_')[1][0].replace('남','1').replace('여','2')
        region = rows
        age = test[cols][rows]
        
        df2 = pd.DataFrame({'region' : [region], 'yyyy' : [yyyy], 'mm' : [mm], 'sex': [sex], 'avg_age': [age]})
        avg_age_frm = pd.concat([avg_age_frm, df2], ignore_index=True)

# region 기준 오름차순 정렬
avg_age_frm = avg_age_frm.sort_values(by=['region','yyyy','mm']).reset_index(drop=True)

display(avg_age_frm.head(10))
print(avg_age_frm.shape)
```
